## Data Analysis

```
df1 = spark.read.format("csv").option("header", "true").load("dbfs:/FileStore/shared_uploads/mkovvuri@gmu.edu/Spring_2022_RM_Survey_Results_100-6.csv")
display(df1)
```

```
display(df1.groupBy("Program of study").count().sort('count'))
```

```
pdf = df1.groupBy("BS field of study").count().sort(desc('count')).toPandas()
pdf['BS field of study'] = pdf['BS field of study'].replace(['Other STEM (Science, Technology, Engineering, Math) majors except for CS/IT'],'Other Stem')
cnt_srs = pdf
trace = go.Scatter(
    x=pdf['BS field of study'],y=pdf['count'],mode='markers',
    marker=dict(sizemode = 'diameter',sizeref = 1,
        size = 50,color = pdf['count'],colorscale='Portland',showscale=True),)
layout = go.Layout(title='Formal education Distribution')

data = [trace]
fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename="Formal education")
```

```
pdf = df1.groupBy("Class format").count().sort(desc('count')).toPandas()
cnt_srs = pdf
trace = go.Scatter(
    x=pdf['Class format'],y=pdf['count'],mode='markers',
    marker=dict(sizemode = 'diameter',sizeref = 1,
        size = 50,color = pdf['count'],colorscale='Portland',showscale=True),)
layout = go.Layout(title='Class Format Distribution')
data = [trace]
fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename="Class format")
```
